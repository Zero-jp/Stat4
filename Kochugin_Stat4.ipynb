## Задание 1

Приведение данных к приемлемому виду.


```python
import pandas as pd
df = pd.read_excel('sample_data/data.xls')
df.dropna(axis = 0, inplace=True)
df.drop(1, inplace=True)
df.columns = ['Кол-во яиц', '1', '2', '3', '4', '5 или более']
df.index = df.index - 2
df
```

Найдём интересующие нас суммы.

```python
df["total"] = df.iloc[:, 1:].sum(axis=1)
df.loc[2] = df.sum(axis=0)
df
```

Частоты:

```python
chast_1_1 = df['1'][0]/df['total'][0]
chast_1_2 = df['2'][0]/df['total'][0]
chast_1_3 = df['3'][0]/df['total'][0]
chast_1_4 = df['4'][0]/df['total'][0]
chast_1_5 = df['5 или более'][0]/df['total'][0]
chast_2_1 = df['1'][1]/df['total'][1]
chast_2_2 = df['2'][1]/df['total'][1]
chast_2_3 = df['3'][1]/df['total'][1]
chast_2_4 = df['4'][1]/df['total'][1]
chast_2_5 = df['5 или более'][1]/df['total'][1]
```

Сумма Пирсона:

```python
weight = df['total'][0]*df['total'][1]
pirs_sum = weight*(1/df['1'][2]*(chast_1_1-chast_2_1)**2+1/df['2'][2]*(chast_1_2-chast_2_2)**2
                   +1/df['3'][2]*(chast_1_3-chast_2_3)**2+1/df['4'][2]*(chast_1_4-chast_2_4)**2+1/df['5 или более'][2]*(chast_1_5-chast_2_5)**2)
pirs_sum
```

Пусть критическое значение = 9.488. Тогда:

```python
pirs_crit = 9.488
if pirs_sum < pirs_crit:
  print('Нет оснований отвергнуть гипотезу об однородности.')
else:
  print('Есть основания отвергнуть гипотезу об однородности.')
```

Разделим полученные данные на две таблицы по выборкам.

Теперь обработаем данные и составим таблицу по колличеству значений попавших в тот, или иной интервал.

Пусть будут следующие интервалы:

[0:2],[2:4],[4:6],[6:8],[8:10]

```python
def fetch_match_times(x, y, df_name):
  out = 0
  for rowIndex, row in df_name.iterrows():
    for columnIndex, value in row.items():
      if ((value < y) & (value > x)):
        out += 1
  return out
df_data = pd.DataFrame(columns=['selections\ranges', '[0:2]','[2:4]','[4:6]','[6:8]','[8:10]'])
df_data.loc[0] = ['df_1', fetch_match_times(0,2, df_1), fetch_match_times(2,4, df_1),
                  fetch_match_times(4,6, df_1), fetch_match_times(6,8, df_1), fetch_match_times(8,10, df_1)]
df_data.loc[1] = ['df_2', fetch_match_times(0,2, df_2), fetch_match_times(2,4, df_2),
                  fetch_match_times(4,6, df_2), fetch_match_times(6,8, df_2), fetch_match_times(8,10, df_2)]
df_data
```

Теперь всё сводится к первой задаче. А именно:
*   Нахождение сумм по строкам и столбцам
*   Нахождение суммы Пирсона
*   Сравнение суммы Пирсона с критическим значением(9.488)



```python
df_data["total"] = df_data.iloc[:, 1:].sum(axis=1)
df_data.loc[2] = df_data.sum(axis=0)
print(df_data)

chast_1_1 = df_data['[0:2]'][0]/df_data['total'][0]
chast_1_2 = df_data['[2:4]'][0]/df_data['total'][0]
chast_1_3 = df_data['[4:6]'][0]/df_data['total'][0]
chast_1_4 = df_data['[6:8]'][0]/df_data['total'][0]
chast_1_5 = df_data['[8:10]'][0]/df_data['total'][0]
chast_2_1 = df_data['[0:2]'][1]/df_data['total'][1]
chast_2_2 = df_data['[2:4]'][1]/df_data['total'][1]
chast_2_3 = df_data['[4:6]'][1]/df_data['total'][1]
chast_2_4 = df_data['[6:8]'][1]/df_data['total'][1]
chast_2_5 = df_data['[8:10]'][1]/df_data['total'][1]

weight = df_data['total'][0]*df_data['total'][1]
pirs_sum = weight*(1/df_data['[0:2]'][2]*(chast_1_1-chast_2_1)**2+1/df_data['[2:4]'][2]*(chast_1_2-chast_2_2)**2
                   +1/df_data['[4:6]'][2]*(chast_1_3-chast_2_3)**2+1/df_data['[6:8]'][2]*(chast_1_4-chast_2_4)**2+1/df_data['[8:10]'][2]*(chast_1_5-chast_2_5)**2)
print('Сумма Пирсона:', pirs_sum)

pirs_crit = 9.488
if pirs_sum < pirs_crit:
  print('Нет оснований отвергнуть гипотезу об однородности.')
else:
  print('Есть основания отвергнуть гипотезу об однородности.')
```
